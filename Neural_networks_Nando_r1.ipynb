# Dataset

I will try to train the neural network to learn 1 wave of a sine function:


```
import numpy as np
from matplotlib import pyplot as plt
```

```
N=5000
x=np.linspace(0,10,N)
sigma=0.1
noise=np.random.normal(0,sigma,N)
y=np.sin(x)+noise
plt.plot(x,y);
```

# Algorithm

This is the algorithm:

image.png

```
#1. Random initialization of thetas.

num_thet=7
sigma=0.1

θs=np.random.normal(0,sigma,num_thet)


#For loop:

num_steps=100000
η=0.00001          #learning rate

for i in range(num_steps):
  y_hat=θs[4]+θs[5]/(1+np.exp(-θs[0]-θs[1]*x))+θs[6]/(1+np.exp(-θs[3]-θs[2]*x))
  y_y_hat=y-y_hat
  MSE=y_y_hat.T.dot(y_y_hat)
  print('step:',i)
  print('MSE:%.5f'%MSE)
  u11=θs[0]+θs[1]*x
  u12=θs[3]+θs[2]*x
  o11=1/(1+np.exp(-u11))
  o12=1/(1+np.exp(-u12))
  dyt5=np.ones(len(x))
  dyt6=o11
  dyt7=o12
  dyt3=θs[6]*o12*(1-o12)*x
  dyt2=θs[5]*o11*(1-o11)*x
  dyt1=θs[5]*o11*(1-o11)
  dyt4=θs[6]*o12*(1-o12)
  y_hat_prime=np.array([dyt1,dyt2,dyt3,dyt4,dyt5,dyt6,dyt7])
  θs+=η*y_hat_prime.dot(y_y_hat)





```

**Let's see if the neural net has learned anything**

```
y_pred=y_hat=θs[4]+θs[5]/(1+np.exp(-θs[0]-θs[1]*x))+θs[6]/(1+np.exp(-θs[3]-θs[2]*x))
```

```
plt.scatter(x,y)
plt.scatter(x,y_pred,color='r',alpha=0.09);
```
